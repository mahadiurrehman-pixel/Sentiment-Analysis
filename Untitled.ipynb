```python
import pandas as pd
from nltk.tokenize import word_tokenize 
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
import re 
from sklearn.metrics import accuracy_score  
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
import pickle 

en_stop = set(stopwords.words("english"))
lemmatizer = WordNetLemmatizer()

def clean_text(text):
    text = re.sub(r'^[^-]*-\s*', '', text)
    text = text.lower()
    text = re.sub(r'[^\w\s]', '', text)

    tokens = text.split()  # Fast
    tokens = [word for word in tokens if word not in en_stop]
    tokens = [lemmatizer.lemmatize(word) for word in tokens]

    return " ".join(tokens)
```

```python
data=pd.read_csv("IMDB Dataset.csv")
data.head()
```

```python
data.info()
```

```python
data['clean_text']=data['review'].apply(clean_text)
```

```python
X = data['clean_text']
y = data['sentiment']
```

```python
count_vec=CountVectorizer()
Tfidf = TfidfVectorizer()
count_vec_fit=count_vec.fit_transform(X)
Tfidf_fit = Tfidf.fit_transform(X)
```

```python
countX_train,countX_test,county_train,county_test = train_test_split(count_vec_fit,y,test_size=0.2,random_state=42)
tfX_train,tfX_test,tfy_train,tfy_test = train_test_split(Tfidf_fit,y,test_size=0.2,random_state=42)
```

```python
lr_count= LogisticRegression().fit(countX_train,county_train)
lr_tf=LogisticRegression().fit(tfX_train,tfy_train)
y_lr_count_predict= lr_count.predict(countX_test)
y_lr_tf_predict = lr_tf.predict(tfX_test)
naive_count=MultinomialNB().fit(countX_train,county_train)
naive_tf=MultinomialNB().fit(tfX_train,tfy_train)
y_naive_count_predict= lr_count.predict(countX_test)
y_naive_tf_predict = lr_tf.predict(tfX_test)
```

```python
print("Logistic Regression + Count:",accuracy_score(county_test,y_lr_count_predict))
print("Logistic Regression + TF-IDF:",accuracy_score(tfy_test,y_lr_tf_predict))
print("Naive Bayes + Count:",accuracy_score(county_test,y_naive_count_predict))
print("Naive Bayes + TF-IDF:",accuracy_score(tfy_test,y_naive_tf_predict))
```
